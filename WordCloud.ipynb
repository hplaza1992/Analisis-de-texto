```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 
import numpy as np
stop=open("stopwords.txt","r").read().splitlines()
```

```python
df = pd.read_csv('data.csv')
df = df.drop(columns=['Id_cliente'])
```

```python
df.shape
```

```python
#cambio de nombre 
df2 = df.rename(columns = {'Respuesta texto libre': 'respuesta'}, inplace = False)
```

```python
#Unimos el campo de texto que estaba separado por filas en una sola para analizarlo
text = " ".join(review for review in df2.respuesta)
```

```python
stopwords = set(STOPWORDS)
stopwords.update(stop)
#Generamos la wordcloud
wordcloud = WordCloud(width = 1000, height = 1000,stopwords=stopwords, background_color="white").generate(text)

# pintamos la wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
